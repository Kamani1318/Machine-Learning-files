```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
mobile = pd.read_csv('Mobile_Train.csv')
```

```python
mobile.head()
```

```python
mobile.describe()
```

# Data Analysis

```python
# Check for null values
mobile.isnull().sum()
mobile = mobile.dropna()
```

```python
mobile.info()
```

```python
# Heatmap
corr = mobile.corr()
fig = plt.figure(figsize = (12,5))
r = sns.heatmap(corr,cmap = 'turbo')
r.set_title('Correlation Matrix')
```

```python
x = mobile.iloc[:,:20]
y = mobile.iloc[:,20]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(x,y,test_size = 0.2, random_state = 42)
```

```python
# Only standardise the input data
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(X_train)
x_test = sc.transform(X_test)
```

# SVM

```python
from sklearn.svm import SVC
svm = SVC(C=2,kernel = 'linear')
```

```python
svm.fit(x_train,Y_train)
y_train_predict_svm = svm.predict(x_train)
y_test_predict_svm = svm.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
from sklearn import metrics
```

```python
print('Train Accuracy score: ',accuracy_score(y_train_predict_svm,Y_train))
print('Test Accuracy score: ',accuracy_score(y_test_predict_svm,Y_test))
print('Classification report: ',classification_report(y_test_predict_svm,Y_test))
cm = confusion_matrix(y_test_predict_svm,Y_test)
cm_display = metrics.ConfusionMatrixDisplay(confusion_matrix = cm)
cm_display.plot()
plt.title('Confusion Matrix')
plt.show()
```

# Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB
gb = GaussianNB()
gb.fit(X_train,Y_train)
y_train_predict_gb = gb.predict(X_train)
y_test_predict_gb = gb.predict(X_test)
```

```python
print('Train Accuracy score: ',accuracy_score(y_train_predict_gb,Y_train))
print('Test Accuracy score: ',accuracy_score(y_test_predict_gb,Y_test))
print('Classification report: ',classification_report(y_test_predict_gb,Y_test))
cm2 = confusion_matrix(y_test_predict_gb,Y_test)
cm_display2 = metrics.ConfusionMatrixDisplay(confusion_matrix = cm2)
cm_display.plot(cmap='pink')
plt.title('Confusion Matrix')
plt.show()
```
